```python
%%file example01.py

def add(a, b):
  return a + b


def input_mock(string):
  return "2"


class OurOwnError(Exception):
  pass


def read_number():
  try:
    return int(input("Enter a number: "))
  except ValueError:
    raise OurOwnError("User failed to enter a number")
```

```python
%%file test01.py

import pytest
from unittest.mock import patch

from example01 import add, read_number, OurOwnError

def test_add():
  assert add(1, 2) == 3

@pytest.mark.parametrize("a, b, expected", [
  (1, 2, 3),
  (-1, 3, 2),
  (0, 0, 0)
])
def test_add_parameterized(a, b, expected):
  assert add(a, b) == expected

@patch("builtins.input", return_value="2")
def test_read_number(input_mock):
  assert read_number() == 2

@patch("builtins.input", return_value="not a number")
def test_read_number_when_user_inputs_not_a_number(input_mock):
  try:
    read_number()
  except OurOwnError as e:
    assert e.args[0] == "User failed to enter a number"
  except Exception as e:
    assert False, "Some unexpected error happened"


```
