```python
"""
import 2x generation_data from the raw url
"""
import pandas as pd
url1 = 'https://raw.githubusercontent.com/MainakRepositor/Datasets/master/Solar%20Plant/Plant_1_Generation_Data.csv'
url2 = 'https://raw.githubusercontent.com/MainakRepositor/Datasets/master/Solar%20Plant/Plant_2_Generation_Data.csv'

df1 = pd.read_csv(url1,index_col=0,parse_dates=[0])
df2 = pd.read_csv(url2,index_col=0,parse_dates=[0])

df1.head(5)
df2.head(5)
```

```python
"""
concordinate two dataframes
"""
df = pd.concat([df1,df2])
df
```

```python
#describe and head of dataframe
df.describe()
```

```python
df.head(10)
```

```python
#SOURCE_KEY that produced the highest DAILY_YIELD
df_max = df.iloc[df[['DAILY_YIELD']].max()]
df_max['SOURCE_KEY']
```

```python
#mean() of `DAILY_YIELD` per plant
df_group = df.groupby('PLANT_ID')
df_group['DAILY_YIELD'].mean()
```

```python
#day that produced the biggest `DAILY_YIELD` for each plant
df_group['DAILY_YIELD'].idxmax()
```

```python
#daily production of energy per energy plant
import plotly.express as px
#df_agg = df('DAILY_YIELD')
plot = px.bar(df,
              x='PLANT_ID',
              y='TOTAL_YIELD',
              title='Total energy per plant',
              template='simple_white'
              )
plot.show()
```
